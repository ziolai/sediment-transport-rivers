# Mathematical Modeling in Civil Engineering 

## Periodic Hydrodynamic Loading of Porous structures 

<b>Model Development and Analytical Solution</b> See journal paper <i>AN ALTERNATIVE PROCESS-BASED APPROACH TO
PREDICTING THE RESPONSE OF WATER-SATURATED POROUS MEDIA TO HARMONIC HYDRODYNAMIC LOADS</i> by Van Damme and Den Ouden-van der Horst, in Journal of Porous Media, 27(1):69–91, 2024. 

<b>Numerical Validation</b> See TU Delft Master thesis [Compressible vs. incompressible pore water in fully-saturated poroelastic soil](https://repository.tudelft.nl/record/uuid:e7a3b318-59a0-4728-8575-f7957e475302) by Femke Klein with corresponding [code](https://github.com/fpmklein/Compressible-vs.-incompressible-pore-water-in-fully-saturated-poroelastic-soil).
