## Section 3: Extension With Acceleration Terms  

It is hypothesized that inclusion of the acceleration terms leads to a solution which is no longer vorticity free. The displacements are either induced by a volumetric strain or by a vorticity. A solution which is both vorticity free and does not give any volumetric strain seems to become dependent on the choice of whether to include the influence of the advective terms or not. When only the local acceleration term is accounted for, then the acceleration due to displacements which are neither vorticity nor volumetric strain induced could be counteracted by the homogeneous solution of the pressure.  This would then result in an extra boundary condition which is needed to solve the dynamic equations. When the advective terms are accounted for, the displacement vector which neither gives a volumetric strain or a vorticity seems implausible and the homogeneous solution of the pressure must become 0 again. The total displacement vector would then consist of a volumetric strain induced deformation part and a vorticity induced deformation part. The hypothesis is that each time step the divergence of the vorticy induced deformation vector must be 0 and the curl of the volumetric strain induced deformation part must also be 0. This may aid in finding a solution.  

The question is about verifying these hypothesis by extending the numerical solution Ref-2 to include all acceleration terms.
